```python
#How do i filter rows of a pandas Dataframe by column value?
```

```python
movies=pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies.shape
```

```python
#selecting rows where duration(a column)is greater than 200 mins

booleans=[]
for length in movies.duration:
    if length>=200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
len(booleans)
#same length of data frame
```

```python
#creating series
is_long=pd.Series(booleans)
```

```python
is_long.head()
```

```python
movies[is_long]
```

```python
#instead of writing for loop
is_long=movies.duration >=200
is_long.head()
```

```python
movies[is_long]
```

```python
#another method
movies[movies.duration >=200]
```

```python
#To get only one data series instead of a complete data frame - example-genre
movies[movies.duration >=200].genre
```

```python
#Or use loc method
movies.loc[movies.duration >=200].genre
```

```python
#How do i apply multiple filter criteria to a pandas Dataframe?
movies=pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
#Filtering out two criterias - using or & and operaters
#Can't use 'and' need to use &
movies[(movies.duration >=200) & (movies.genre == 'Crime')]
```

```python
#For 'or' use |
movies[(movies.duration >=200)  | (movies.genre == 'Crime')]
```

```python
#passing too many conditions
movies[(movies.genre=='Drama')  | (movies.genre == 'Crime')  | (movies.genre == 'Action')]
```

```python
movies[movies.genre.isin(['Crime','Drama','Action'])]
```
