```python
#Renaming columns in data frame
```

```python
ufo=pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head()
```

```python
ufo.columns
```

```python
#METHOD 1

ufo.rename(columns= {'Colors Reported':'Colors_Reported','Shape Reported':'Shape_Reported'},inplace=True)
#Passing in a dictionary as key value pairs using argument columns
```

```python
ufo.head()
```

```python
#making column names in lowercase
ufo_cols=['city','colors reported','shape reported','state','time']
```

```python
#METHOD 2
ufo.columns=ufo_cols
```

```python
ufo.head()
```

```python
#METHOD 3
#overwriting column names while reading the files
ufo=pd.read_csv('http://bit.ly/uforeports',names=ufo_cols,header=0)
#header is 0 as you are overwriting the column names
```

```python
ufo.head()
```

```python
#METHOD 4 -BEST
ufo.columns=ufo.columns.str.replace(' ','_')
```
