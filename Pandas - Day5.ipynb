# How do i select multiple rows and columns from a pandas data frame

```python
import pandas as pd
```

```python
ufo=pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head(3)
```

```python
#loc is used to select rows and columns by label
#example- selecting first row 
#loc[row label,column label]
ufo.loc[0,:]
```

```python
#getting multiple rows-pass a list
ufo.loc[[0,1,2],:]
```

```python
ufo.loc[0:2,:]
#includes 2nd row not as a range
```

```python
ufo.loc[:,['City','State']]
```

```python
#specifying as range
ufo.loc[:,'City':'State']
```

```python
ufo.loc[0:2,'City':'State']
```

```python
#can acheive the same thing using drop method
ufo.head(3).drop('Time',axis=1)
```

```python
ufo[ufo.City=='Oakland']
```

```python
#can do the above usin loc
ufo.loc[ufo.City=='Oakland',:]
```

```python
ufo[ufo.City=='Oakland'].State
```

```python
ufo.loc[ufo.City=='Oakland','State']
#using loc is best as it doesn't cause problems
```

# iloc -filtering rows and columns by integer positions

```python
ufo.iloc[:,[0,3]]
```

```python
ufo.loc[:,[0,3]]
#will show error as loc can't filter columns and rows using integer values
```

```python
ufo.columns
```

```python
ufo.iloc[:,0:4]
#exclusive for end integer - excludes 4 th column
```

```python
ufo.iloc[0:3,:]
#did not include row 3 -big difference between loc and iloc
```

```python
ufo[['City','State']]
```

```python
#instead use loc
ufo.loc[:,['City','State']]
```

```python
ufo[0:2]
#confusing as doesn't specify if its for rows or columns
```

```python
ufo.iloc[0:2,:]
```

# when shoud i use 'inplace' parameter in pandas?

```python
ufo.shape
```

```python
ufo.head()
```

```python
ufo.drop('City',axis=1)
```

```python
ufo.head()
#even though we have removed city using drop city is still there
#Because drop has default inplace=False
```

```python
#to remove permanently-
ufo.drop('City',axis=1,inplace=True)
```

```python
ufo.head()
```

```python
ufo.dropna(how='any').shape
```

```python
ufo.shape
```

```python
ufo.set_index('Time',inplace=True)
```

```python
#can use assignment operater but the above method is perfect
ufo=ufo.set_index('Time')
```
